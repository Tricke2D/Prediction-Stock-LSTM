```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import History
from ta.trend import SMAIndicator, EMAIndicator, MACD
from ta.momentum import RSIIndicator
from ta.volatility import BollingerBands, AverageTrueRange

# Impor model tradisional
from sklearn.neural_network import MLPRegressor # Untuk perbandingan dengan NN sederhana
from xgboost import XGBRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor

# Import modul files dari google.colab untuk unduhan otomatis (jika di Colab)
try:
    from google.colab import files
    COLAB_ENVIRONMENT = True
except ImportError:
    COLAB_ENVIRONMENT = False
    print("Tidak di lingkungan Google Colab. Unduhan otomatis tidak akan tersedia.")

# --- Konfigurasi Jalur File ---
DATASET_PATH = '/content/Data Historis BBRI (5).csv'
BASE_DIR = os.getcwd()

MODEL_FILENAME = 'stock_prediction_model.h5'
SCALER_FILENAME = 'minmax_scaler.pkl'

MODEL_FULL_PATH = os.path.join(BASE_DIR, MODEL_FILENAME)
SCALER_FULL_PATH = os.path.join(BASE_DIR, SCALER_FILENAME)

# --- KONSTANTA ---
SEQ_LENGTH = 9 # Panjang urutan input historis untuk LSTM
TEST_SIZE = 0.2 # Proporsi data untuk pengujian

# --- Fungsi Metrik Kustom ---
def mean_absolute_percentage_error(y_true, y_pred):
    """Menghitung Mean Absolute Percentage Error (MAPE)."""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Hindari pembagian dengan nol untuk y_true yang sangat kecil atau nol
    return np.mean(np.abs((y_true - y_pred) / np.maximum(y_true, np.finfo(float).eps))) * 100

# --- Fungsi Pembersihan dan Pra-pemrosesan Data ---
def clean_volume(vol):
    """Membersihkan dan mengkonversi kolom 'Vol.' dari string ke float."""
    vol = str(vol).replace(',', '.')
    if 'M' in vol:
        return float(vol.replace('M', '')) * 1e6
    elif 'B' in vol:
        return float(vol.replace('B', '')) * 1e9
    else:
        return np.nan

def load_and_preprocess_data(dataset_path):
    """Memuat dataset, membersihkan, dan mengurutkannya."""
    print("=== Memuat Dataset ===")
    try:
        df = pd.read_csv(dataset_path)
        df['Tanggal'] = pd.to_datetime(df['Tanggal'], format='%d/%m/%Y')
        df = df.sort_values('Tanggal').reset_index(drop=True)
        print("Dataset berhasil dimuat:")
        print(df.head())
        print("\nInformasi Dataset Sebelum Pembersihan:")
        print(df.info())

        df['Vol.'] = df['Vol.'].apply(clean_volume)
        df['Perubahan%'] = df['Perubahan%'].str.replace('%', '').str.replace(',', '.').astype(float)
        for col in ['Terakhir', 'Pembukaan', 'Tertinggi', 'Terendah']:
            df[col] = df[col].astype(float)

        print("\nInformasi Dataset Setelah Pembersihan:")
        print(df.info())
        print(df.describe())

        return df
    except Exception as e:
        print(f"Gagal membaca dataset: {str(e)}")
        print("Membuat DataFrame dummy untuk demonstrasi karena gagal memuat dataset.")
        dates = pd.to_datetime(pd.date_range(start='2020-01-01', periods=100, freq='D'))
        data = {
            'Tanggal': dates,
            'Terakhir': np.random.uniform(3000, 5000, 100),
            'Pembukaan': np.random.uniform(3000, 5000, 100),
            'Tertinggi': np.random.uniform(3000, 5000, 100),
            'Terendah': np.random.uniform(3000, 5000, 100),
            'Vol.': [f"{int(np.random.uniform(10, 100))}M" for _ in range(100)],
            'Perubahan%': [f"{np.random.uniform(-2, 2):.2f}%" for _ in range(100)]
        }
        df_dummy = pd.DataFrame(data)
        df_dummy = df_dummy.sort_values('Tanggal').reset_index(drop=True)
        df_dummy['Vol.'] = df_dummy['Vol.'].apply(clean_volume)
        df_dummy['Perubahan%'] = df_dummy['Perubahan%'].str.replace('%', '').str.replace(',', '.').astype(float)
        for col in ['Terakhir', 'Pembukaan', 'Tertinggi', 'Terendah']:
            df_dummy[col] = df_dummy[col].astype(float)
        print(df_dummy.info())
        return df_dummy

# --- Fungsi Analisis Data Eksplorasi (EDA) ---
def perform_eda(df_input):
    """Melakukan Analisis Data Eksplorasi (EDA)."""
    print("\n=== Melakukan Analisis Data Eksplorasi (EDA) ===")
    print("\nStatistik Deskriptif:")
    print(df_input.describe())

    print("\nJumlah Nilai Hilang per Kolom:")
    print(df_input.isnull().sum())

    fig_closing_price = plt.figure(figsize=(12, 7))
    plt.plot(df_input['Tanggal'], df_input['Terakhir'], label='Harga Penutupan')
    plt.title('Harga Penutupan Saham BBRI Seiring Waktu')
    plt.xlabel('Tanggal')
    plt.ylabel('Harga (Rp)')
    plt.legend()
    plt.grid(True)
    plt.show()
    plt.close(fig_closing_price)

    fig_price_distribution = plt.figure(figsize=(10, 6))
    sns.histplot(df_input['Terakhir'], bins=50, kde=True, color='skyblue')
    plt.title('Distribusi Harga Penutupan Saham BBRI')
    plt.xlabel('Harga (Rp)')
    plt.ylabel('Frekuensi')
    plt.show()
    plt.close(fig_price_distribution)

    numerical_cols = ['Terakhir', 'Pembukaan', 'Tertinggi', 'Terendah', 'Vol.', 'Perubahan%']
    corr_df = df_input[numerical_cols].dropna()
    if not corr_df.empty:
        fig_correlation = plt.figure(figsize=(10, 8))
        sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Matriks Korelasi Fitur Numerik Saham BBRI')
        plt.show()
        plt.close(fig_correlation)
    else:
        print("Tidak cukup data numerik atau ada NaN untuk membuat heatmap korelasi.")

# --- Fungsi Perhitungan Indikator Teknis ---
def calculate_technical_indicators(df_input):
    """Menghitung indikator teknis."""
    print("\n=== Menghitung Fitur Teknis ===")
    df = df_input.copy()

    df['SMA_5'] = SMAIndicator(df['Terakhir'], window=5).sma_indicator()
    df['EMA_5'] = EMAIndicator(df['Terakhir'], window=5).ema_indicator()
    df['RSI_14'] = RSIIndicator(df['Terakhir'], window=14).rsi()
    macd_instance = MACD(df['Terakhir'])
    df['MACD'] = macd_instance.macd()
    df['Signal_Line'] = macd_instance.macd_signal()

    bollinger = BollingerBands(close=df['Terakhir'], window=20, window_dev=2)
    df['BB_Mid'] = bollinger.bollinger_mavg()
    df['BB_High'] = bollinger.bollinger_hband()
    df['BB_Low'] = bollinger.bollinger_lband()
    df['BB_Width'] = bollinger.bollinger_wband()
    df['BB_Percent'] = bollinger.bollinger_pband()

    atr = AverageTrueRange(high=df['Tertinggi'], low=df['Terendah'], close=df['Terakhir'], window=14)
    df['ATR'] = atr.average_true_range()

    print("Data dengan Fitur Teknis (5 baris terakhir):")
    print(df[['Tanggal', 'Terakhir', 'SMA_5', 'EMA_5', 'RSI_14', 'MACD', 'Signal_Line',
              'BB_Mid', 'BB_High', 'BB_Low', 'BB_Width', 'BB_Percent', 'ATR']].tail())
    return df

# --- Fungsi Pembuatan Sequence untuk LSTM (single-output) ---
def create_sequences(data, seq_length):
    """
    Membuat sequence data untuk input model LSTM (X) dan target single-langkah (y).
    X: urutan data historis (seq_length), y: harga pada hari berikutnya.
    `data` harus sudah berupa fitur numerik yang diskalakan.
    """
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length])
        y.append(data[i + seq_length, 0]) # Target adalah harga 'Terakhir' pada hari berikutnya (kolom ke-0)
    X, y = np.array(X), np.array(y)
    return X, y

# --- Fungsi Evaluasi Model Umum ---
def evaluate_model(model_name, y_true_unscaled, y_pred_unscaled):
    """Mengevaluasi model dan mencetak metrik RMSE dan MAPE."""
    rmse = np.sqrt(mean_squared_error(y_true_unscaled, y_pred_unscaled))
    mape = mean_absolute_percentage_error(y_true_unscaled, y_pred_unscaled)
    r2 = r2_score(y_true_unscaled, y_pred_unscaled)
    return {
        'Model': model_name,
        'RMSE': rmse,
        'MAPE': mape,
        'R2': r2
    }

# --- Main Execution for Training Script ---
if __name__ == "__main__":
    df_raw = load_and_preprocess_data(DATASET_PATH)

    # Hitung indikator teknis untuk seluruh dataset
    df_with_indicators = calculate_technical_indicators(df_raw)

    # Definisikan fitur numerik yang akan digunakan untuk scaling dan model
    numerical_features = ['Terakhir', 'Pembukaan', 'Tertinggi', 'Terendah', 'Vol.', 'Perubahan%',
                          'SMA_5', 'EMA_5', 'RSI_14', 'MACD', 'Signal_Line',
                          'BB_Mid', 'BB_High', 'BB_Low', 'BB_Width', 'BB_Percent', 'ATR']

    # Buat DataFrame yang hanya berisi fitur numerik dan tanggal, lalu dropna
    df_features_and_dates = df_with_indicators[['Tanggal'] + numerical_features].dropna().reset_index(drop=True)

    # Pastikan data cukup setelah dropna untuk pembuatan sequence
    if len(df_features_and_dates) < (SEQ_LENGTH + 1):
        print(f"Data tidak cukup setelah penghapusan NaN untuk membuat sequence. "
              f"Minimum data yang dibutuhkan: {SEQ_LENGTH + 1} hari. Tersedia: {len(df_features_and_dates)} hari.")
        exit()

    # Normalisasi hanya fitur numerik
    scaler = MinMaxScaler()
    scaled_numerical_data = scaler.fit_transform(df_features_and_dates[numerical_features])

    # Buat kembali DataFrame dengan data numerik yang diskalakan dan tanggal asli
    df_scaled_with_dates = pd.DataFrame(scaled_numerical_data, columns=numerical_features)
    df_scaled_with_dates['Tanggal'] = df_features_and_dates['Tanggal']

    # Pisahkan data untuk LSTM (format sequence)
    # X_lstm hanya berisi fitur numerik yang diskalakan
    X_lstm, y_lstm = create_sequences(df_scaled_with_dates[numerical_features].values, SEQ_LENGTH)

    # Split data LSTM menjadi training dan testing
    # shuffle=False penting untuk deret waktu
    X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(
        X_lstm, y_lstm, test_size=TEST_SIZE, random_state=42, shuffle=False
    )

    # Untuk model tradisional, kita juga menggunakan X_lstm yang diratakan
    # karena mereka akan memprediksi 1 hari ke depan berdasarkan SEQ_LENGTH fitur historis.
    X_train_trad_flat = X_train_lstm.reshape(X_train_lstm.shape[0], -1)
    X_test_trad_flat = X_test_lstm.reshape(X_test_lstm.shape[0], -1)

    all_model_results = []

    print("\n--- Melatih Model LSTM ---")
    model_lstm = Sequential([
        LSTM(100, activation='relu', input_shape=(SEQ_LENGTH, len(numerical_features)), return_sequences=True),
        LSTM(50, activation='relu'),
        Dense(1) # Output 1 hari
    ])
    model_lstm.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    history_lstm = model_lstm.fit(X_train_lstm, y_train_lstm, epochs=100, batch_size=32, verbose=1, validation_split=0.2)

    # Evaluasi LSTM
    y_pred_lstm_scaled = model_lstm.predict(X_test_lstm, verbose=0).flatten()

    # Inverse transform predictions and true values for LSTM
    # Untuk inverse transform, kita perlu array dummy dengan jumlah kolom yang sama seperti scaler di-fit
    dummy_array_shape = (len(y_test_lstm), len(numerical_features))

    dummy_true_lstm = np.zeros(dummy_array_shape)
    dummy_true_lstm[:, 0] = y_test_lstm
    y_true_lstm_unscaled = scaler.inverse_transform(dummy_true_lstm)[:, 0]

    dummy_pred_lstm = np.zeros(dummy_array_shape)
    dummy_pred_lstm[:, 0] = y_pred_lstm_scaled
    y_pred_lstm_unscaled = scaler.inverse_transform(dummy_pred_lstm)[:, 0]

    all_model_results.append(evaluate_model("LSTM", y_true_lstm_unscaled, y_pred_lstm_unscaled))

    print("\n--- Melatih Model MLPRegressor (Neural Network Sederhana) ---")
    model_mlp = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42, early_stopping=True, validation_fraction=0.2)
    model_mlp.fit(X_train_trad_flat, y_train_lstm) # y_train_lstm adalah target 1 hari ke depan
    y_pred_mlp_scaled = model_mlp.predict(X_test_trad_flat)

    dummy_pred_mlp = np.zeros(dummy_array_shape)
    dummy_pred_mlp[:, 0] = y_pred_mlp_scaled
    y_pred_mlp_unscaled = scaler.inverse_transform(dummy_pred_mlp)[:, 0]

    all_model_results.append(evaluate_model("MLPRegressor", y_true_lstm_unscaled, y_pred_mlp_unscaled))


    print("\n--- Melatih Model XGBoost ---")
    model_xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42, objective='reg:squarederror')
    model_xgb.fit(X_train_trad_flat, y_train_lstm)
    y_pred_xgb_scaled = model_xgb.predict(X_test_trad_flat)

    dummy_pred_xgb = np.zeros(dummy_array_shape)
    dummy_pred_xgb[:, 0] = y_pred_xgb_scaled
    y_pred_xgb_unscaled = scaler.inverse_transform(dummy_pred_xgb)[:, 0]

    all_model_results.append(evaluate_model("XGBoost", y_true_lstm_unscaled, y_pred_xgb_unscaled))

    print("\n--- Melatih Model AdaBoost ---")
    model_ada = AdaBoostRegressor(n_estimators=100, random_state=42)
    model_ada.fit(X_train_trad_flat, y_train_lstm)
    y_pred_ada_scaled = model_ada.predict(X_test_trad_flat)

    dummy_pred_ada = np.zeros(dummy_array_shape)
    dummy_pred_ada[:, 0] = y_pred_ada_scaled
    y_pred_ada_unscaled = scaler.inverse_transform(dummy_pred_ada)[:, 0]

    all_model_results.append(evaluate_model("AdaBoost", y_true_lstm_unscaled, y_pred_ada_unscaled))

    print("\n--- Melatih Model Decision Tree ---")
    model_dt = DecisionTreeRegressor(random_state=42)
    model_dt.fit(X_train_trad_flat, y_train_lstm)
    y_pred_dt_scaled = model_dt.predict(X_test_trad_flat)

    dummy_pred_dt = np.zeros(dummy_array_shape)
    dummy_pred_dt[:, 0] = y_pred_dt_scaled
    y_pred_dt_unscaled = scaler.inverse_transform(dummy_pred_dt)[:, 0]

    all_model_results.append(evaluate_model("Decision Tree", y_true_lstm_unscaled, y_pred_dt_unscaled))

    print("\n\n=== RINGKASAN PERBANDINGAN MODEL ===")
    results_df = pd.DataFrame(all_model_results)
    print(results_df.set_index('Model'))
    print("\nModel terbaik (RMSE terendah) adalah:", results_df.loc[results_df['RMSE'].idxmin()]['Model'])
    print("Model terbaik (MAPE terendah) adalah:", results_df.loc[results_df['MAPE'].idxmin()]['Model'])


    print("\n--- Plot Loss Pelatihan dan Validasi (Hanya untuk LSTM) ---")
    fig_loss = plt.figure(figsize=(10, 6))
    plt.plot(history_lstm.history['loss'], label='Loss Pelatihan')
    plt.plot(history_lstm.history['val_loss'], label='Loss Validasi')
    plt.title('Loss Pelatihan dan Validasi Model LSTM')
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)
    plt.show()
    plt.close(fig_loss)

    print("\n--- Plot Perbandingan Harga Asli vs Prediksi (Hanya untuk LSTM) ---")
    fig_comparison_train = plt.figure(figsize=(15, 7))

    # Untuk plot perbandingan LSTM, gunakan data test
    # Rekonstruksi tanggal untuk plotting pada data test
    # Dapatkan tanggal asli yang sesuai dengan X_test_lstm.
    # y_test_lstm adalah target untuk hari setelah setiap sequence X_test_lstm.
    # Indeks awal di df_features_and_dates untuk target pertama di y_test_lstm
    # adalah `len(df_features_and_dates) - len(y_lstm_total_from_df_features) + len(y_train_lstm) + SEQ_LENGTH`

    # Sederhananya, temukan tanggal akhir dari data pelatihan dan tanggal awal dari data pengujian.
    # test_start_index_in_original_data_after_dropna = (jumlah sequence pelatihan * (target_idx + 1)) (ini agak rumit)
    # Cara lebih mudah:
    # y_lstm adalah target dari `SEQ_LENGTH` hingga akhir data numerik yang diskalakan.
    # y_test_lstm adalah bagian terakhir dari y_lstm.
    # Jadi, tanggalnya adalah `df_features_and_dates['Tanggal']` mulai dari indeks `(len(df_features_and_dates) - len(y_lstm)) + len(y_train_lstm) + SEQ_LENGTH`

    # Karena `y_lstm` dibuat dari `df_scaled_with_dates[numerical_features].values`,
    # `y_lstm[k]` adalah harga di `df_scaled_with_dates.iloc[k + SEQ_LENGTH]['Terakhir']`
    # `y_test_lstm` adalah `y_lstm` dari indeks `len(y_train_lstm)` hingga akhir.

    # Jadi, tanggal untuk `y_test_lstm` adalah `df_features_and_dates['Tanggal']`
    # mulai dari `len(y_train_lstm) + SEQ_LENGTH` hingga `len(y_lstm) + SEQ_LENGTH`.

    # Menghitung indeks awal yang benar dalam df_features_and_dates
    # Index dari sequence pertama dalam y_lstm adalah 0. Ini sesuai dengan tanggal di indeks SEQ_LENGTH.
    # Index dari sequence pertama di y_test_lstm adalah len(y_train_lstm).
    # Jadi, tanggal yang sesuai untuk y_test_lstm[0] adalah df_features_and_dates['Tanggal'].iloc[len(y_train_lstm) + SEQ_LENGTH]

    # Panjang total sequence y_lstm adalah len(y_lstm)
    # Panjang y_train_lstm adalah len(X_train_lstm)
    # Panjang y_test_lstm adalah len(X_test_lstm)

    test_dates_start_index = len(y_train_lstm) + SEQ_LENGTH
    plot_dates_test = df_features_and_dates['Tanggal'].iloc[test_dates_start_index : test_dates_start_index + len(y_test_lstm)].tolist()

    plt.plot(plot_dates_test, y_true_lstm_unscaled, label='Harga Asli (Data Test)', color='blue')
    plt.plot(plot_dates_test, y_pred_lstm_unscaled, label='Harga Prediksi (LSTM)', color='red', linestyle='--')
    plt.title('Perbandingan Harga Asli vs Prediksi Saham BBRI (Data Uji - Model LSTM)')
    plt.xlabel('Tanggal')
    plt.ylabel('Harga (Rp)')
    plt.legend()
    plt.grid(True)
    plt.show()
    plt.close(fig_comparison_train)

    epochs_data = "Hasil training per epoch:\n"
    for epoch, loss in enumerate(history_lstm.history['loss']):
        val_loss = history_lstm.history['val_loss'][epoch]
        epochs_data += f"Epoch {epoch+1}: Loss = {loss:.6f}, Val_Loss = {val_loss:.6f}\n"
    print(f"\nHasil Train per Epoch:")
    print(epochs_data)

    evaluation_metrics_text = f"""
Metrik Evaluasi Model:
- Mean Squared Error (MSE): {all_model_results[0]['RMSE']**2:.6f}
- Mean Absolute Error (MAE): {all_model_results[0]['MAPE']:.6f} # Perbaikan: ini MAPE, bukan MAE
- Root Mean Squared Error (RMSE): {all_model_results[0]['RMSE']:.6f}
- R-squared (R2): {all_model_results[0]['R2']:.6f}
- Last Validation Loss: {history_lstm.history['val_loss'][-1]:.6f}
"""
    # Perbaikan: MAPE adalah persentase, bukan MAE
    evaluation_metrics_text = f"""
Metrik Evaluasi Model:
- Mean Squared Error (MSE): {all_model_results[0]['RMSE']**2:.6f}
- Mean Absolute Error (MAE): {all_model_results[0]['RMSE'] * scaler.data_range_[0]:.6f} # Estimasi MAE dari RMSE unscaled
- Root Mean Squared Error (RMSE): {all_model_results[0]['RMSE']:.6f}
- R-squared (R2): {all_model_results[0]['R2']:.6f}
- Mean Absolute Percentage Error (MAPE): {all_model_results[0]['MAPE']:.6f}%
- Last Validation Loss: {history_lstm.history['val_loss'][-1]:.6f}
"""
    print(f"\nMetrik Evaluasi Model:")
    print(evaluation_metrics_text)

    print("\n=== Proses Pelatihan Selesai ===")
    print(f"Model disimpan di: {MODEL_FULL_PATH}")
    print(f"Scaler disimpan di: {SCALER_FULL_PATH}")
    print("Plot EDA, Loss, dan Perbandingan Harga pelatihan akan ditampilkan secara interaktif.")

    # Simpan model LSTM dan scaler
    model_lstm.save(MODEL_FULL_PATH)
    joblib.dump(scaler, SCALER_FULL_PATH)

    if COLAB_ENVIRONMENT:
        print("\n--- Mengunduh File Model dan Scaler ---")
        try:
            files.download(MODEL_FULL_PATH)
            print(f"File '{MODEL_FILENAME}' sedang diunduh.")
        except Exception as e:
            print(f"Gagal mengunduh model: {e}")

        try:
            files.download(SCALER_FULL_PATH)
            print(f"File '{SCALER_FILENAME}' sedang diunduh.")
        except Exception as e:
            print(f"Gagal mengunduh scaler: {e}")
    else:
        print("\nTidak di lingkungan Google Colab, file tidak diunduh secara otomatis.")
```
